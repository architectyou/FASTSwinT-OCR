```python
import os
import shutil
import random

# 원본 이미지가 있는 폴더 경로
source_folder = '/home/pirl/Desktop/OCR/FAST/data/testset/imgs/'

# 이미지를 복사할 대상 폴더 경로
destination_folder = '/home/pirl/Desktop/OCR/FAST/data/1213_random_sample/imgs/'

# 몇 장의 이미지를 랜덤으로 선택할 것인지 설정
num_images_to_select = 100

# 원본 폴더 내의 모든 이미지 파일 리스트 가져오기
image_files = [f for f in os.listdir(source_folder) if f.endswith(('.jpg', '.jpeg', '.png', '.gif', '.JPG', '.JPEG', '.PNG'))]

# 랜덤으로 이미지 파일 선택
selected_images = random.sample(image_files, num_images_to_select)

# 대상 폴더가 없다면 생성
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# 선택한 이미지 파일을 대상 폴더로 복사
for image in selected_images:
    source_path = os.path.join(source_folder, image)
    destination_path = os.path.join(destination_folder, image)
    shutil.copy(source_path, destination_path)

print(f"{num_images_to_select}장의 이미지를 랜덤으로 선택하여 {destination_folder}로 이동했습니다.")
```

```python
import os
import shutil

# 이미지 폴더와 GT 폴더의 경로 설정
image_folder = '/home/pirl/Desktop/OCR/FAST/data/1213_random_sample/imgs/'
gt_folder = '/home/pirl/Desktop/OCR/FAST/data/testset/gts_json/'
destination_folder = '/home/pirl/Desktop/OCR/FAST/data/1213_random_sample/gts_json/'

# GT 폴더 내의 파일 목록 가져오기
gt_files = os.listdir(gt_folder)
if not os.path.isdir(destination_folder):
    os.mkdir(destination_folder)

# GT 파일들을 순환하면서 이미지 폴더에서 일치하는 이미지 파일을 찾고 복사
for gt_file in gt_files:
    gt_path = os.path.join(gt_folder, gt_file)
    
    # GT 파일 이름에서 확장자 제거 및 파일명 추출
    gt_filename_no_ext, _ = os.path.splitext(gt_file)
    
    # 이미지 폴더에서 일치하는 이미지 파일명 생성
    for image_filename in os.listdir(image_folder):
        if gt_filename_no_ext in image_filename:
            image_path = os.path.join(image_folder, image_filename)
            destination_path = os.path.join(destination_folder, gt_file)
            shutil.copy(gt_path, destination_path)
            print(f'복사 완료: {gt_file} -> {destination_path}')

print('작업 완료')
```
